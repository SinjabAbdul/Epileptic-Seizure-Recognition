## Background & Prior Work

Epilepsy is one of the most common neurological diseases globally according to the World Health Organization. Additionally the World Health Organization estimates that, “70% of people living with epilepsy could live seizure-free if properly diagnosed and treated.” There have been many past and ongoing revolutionary efforts to predict as well as efficiently cure people with Epilepsy. Previous work like In Engel, J Jr 2001 research paper, “A proposed diagnostic scheme for people with epileptic seizures and with epilepsy: report of the ILAE Task Force on Classification and Terminology”, aimed to develop a method of a diagnostic scheme in order to categorize people who suffer from epileptic seizures. These categorizations are of two types; first of which is, the seizure type classification which focuses on the characterization of the clinical manifestations of seizures, while the second category is epilepsy classification aims to provide a framework for diagnosing epilepsy syndromes based on clinical and electroencephalographic features. Engel makes note of the importance of gathering details regarding the occurrence of those seizures being activities that occur within both the brain and body at the time before the occurrence of the seizure and during the seizure. The recognition of these categories opened the door for various frameworks of combating epilepsy. With the advancement of technology other research has emerged especially within the field of machine learning. A study conducted in the year 2018 by Yanli Yang, Mengni Zhou, Yan Niu, Conggai Li, Rui Cao, Bin Wang, Pengfei Yan, Yao Ma, Jin Xiang in which they published a paper called, “Epileptic Seizure Prediction Based on Permutation Entropy”, where they explore use of permutation entropy (PE) for predicting epileptic seizures based on intracranial electroencephalogram (iEEG) recordings. The study focuses on tracking the dynamical changes in human brain activity using PE extracted from iEEG signals. The researchers obtained iEEG recordings from 19 patients. They applied preprocessing techniques to the data and extracted PE in a sliding time window. To classify the cerebral state and predict seizures, they implemented a support vector machine. A two-step post-processing method was then used for prediction purposes. This study showed the effectiveness of using PE and SVM for epileptic seizure prediction based on iEEG recordings. The results highlight the potential of this approach in clinical settings, allowing for timely interventions and improved management of epilepsy. A more recent study in 2022 by researchers Milind Natu, Mrinal Bachute, Ketan Kotecha, and Ankit Vidyarthi in their paper, “Review on Epileptic Seizure Prediction: Machine Learning and Deep Learning Approaches” in which examine the various techniques used for seizure prediction, focusing on feature selection and classification, which are the main challenges in developing accurate epilepsy prediction algorithms. The authors categorize the research into three main areas: machine learning and feature extraction methods, artificial neural network-based detection methods, and deep learning methods. Each of the three main areas propose different approaches such as using SVM classifiers to categorize EEG dynamics related to emotions, analyzing spectral power differences for seizure detection, employing Bayesian modeling for multichannel EEG signals, and utilizing empirical mode decomposition (EMD) for detecting prestage seizures. The performance of these methods is evaluated based on parameters like sensitivity, specificity, and accuracy. The article emphasizes the advantages and limitations of each technique and discusses the importance of long-term continuous signal testing to evaluate realistic scenario prediction. It also mentions the availability of publicly available datasets for research purposes. Epilepsy is a chronic disorder of the brain characterized by sudden abnormal discharges of brain neurons, leading to temporary dysfunction. The ability to predict seizures in advance could greatly benefit epilepsy patients by allowing them to take preventive measures and improve their safety and quality of life. However, the potential of PE as a predictive tool for epilepsy remains unclear..We hope to build upon some of the prior work we cited in order to better predict the occurrence of seizures.

## Wavelet Analysis of Trials
Below we begin performing a general wavelet analysis on the EEG trials in the seizure and non-seizure groups, in the sense that we are calculating the wavelet coefficients for each trial using the 'db4' wavelet. The mean and standard deviation of the coefficients are then calculated for each level of decomposition. The code plots the mean and standard deviation of the wavelet coefficients for each level separately for the seizure and non-seizure groups. This allows for visual comparison of the coefficient patterns between the two groups and can provide insights into the differences in brain activity.

```python
import pywt


wavelet = 'db4'

seizure_trials = clean1[clean1['y'] == 1].iloc[:, 1:-1].values
non_seizure_trials = clean1[clean1['y'] == 0].iloc[:, 1:-1].values

seizure_coeffs_all = pywt.wavedec(seizure_trials, wavelet)
seizure_coeffs_mean = [np.mean(level_coeffs, axis=0) for level_coeffs in seizure_coeffs_all]
seizure_coeffs_std = [np.std(level_coeffs, axis=0) for level_coeffs in seizure_coeffs_all]

non_seizure_coeffs_all = pywt.wavedec(non_seizure_trials, wavelet)
non_seizure_coeffs_mean = [np.mean(level_coeffs, axis=0) for level_coeffs in non_seizure_coeffs_all]
non_seizure_coeffs_std = [np.std(level_coeffs, axis=0) for level_coeffs in non_seizure_coeffs_all]

plt.figure(figsize=(10, 6))
num_levels = len(seizure_coeffs_all)
for j in range(num_levels):
    plt.subplot(num_levels, 2, j*2 + 1)
    plt.plot(seizure_coeffs_mean[j])
    plt.title(f'Seizure Group Wavelet Coefficients  Level {j+1} (Mean)')
    plt.xlabel('Time')
    plt.ylabel('Coefficient')

    plt.subplot(num_levels, 2, j*2 + 2)
    plt.plot(seizure_coeffs_std[j])
    plt.title(f'Seizure Group Wavelet Coefficients Level {j+1} (Standard Deviation)')
    plt.xlabel('Time')
    plt.ylabel('Coefficient')

plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
num_levels = len(non_seizure_coeffs_all)
for j in range(num_levels):
    plt.subplot(num_levels, 2, j*2 + 1)
    plt.plot(non_seizure_coeffs_mean[j])
    plt.title(f'Non-Seizure Group Wavelet Coefficients - Level {j+1} (Mean)')
    plt.xlabel('Time')
    plt.ylabel('Coefficient')

    plt.subplot(num_levels, 2, j*2 + 2)
    plt.plot(non_seizure_coeffs_std[j])
    plt.title(f'Non-Seizure Group Wavelet Coefficients Level {j+1} (Standard Deviation)')
    plt.xlabel('Time')
    plt.ylabel('Coefficient')

plt.tight_layout()
plt.show()
```

Now we wanted to dive deeper in our wavelet analysis on the EEG trials in the seizure and non-seizure groups. So we decided to select the first five trials from each group and extracts the signal values. The continuous wavelet transform is applied to each signal using the 'morl' wavelet. The resulting wavelet coefficients are then visualized using an image plot, where the x-axis represents time, the y-axis represents the scale, and the color represents the coefficient magnitude. The purpose is to visualize and compare the wavelet coefficients between the seizure and non-seizure groups, providing insights into the differences in brain activity patterns.

# Spectral Analysis

```python
import numpy as np
import pandas as pd
from scipy.signal import welch
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

def get_spectral_features(series, samprate):
    freqs, psd = welch(series, fs=samprate, nperseg=samprate)
    return pd.Series(psd.sum(axis=0), name='spectral_features')

#nperseg is the value that dictates the size of our window for FFT
#returns a dataframecalled spectral features
```

```python
dataset1 = pd.read_csv('Epileptic Seizure Recognition.csv')
dataset1 = dataset1.iloc[:, 1:179]  # adjust indices to include only EEG data columns
d1samprate = 178  # sample rate for dataset1
```

```python
dataset2_npz = np.load('eeg-predictive_train.npz')
dataset2_array = dataset2_npz['train_signals']
num_samples, num_channels, num_points = dataset2_array.shape
dataset2 = pd.DataFrame(dataset2_array.reshape(num_samples, num_channels * num_points))
d2samprate = 256  # sample rate for dataset2
```

Next, for both datasets we want to take spectral features we analyze and add them to resepective lists. And then we plot them in order to observe the relationship between response magnitude and seizures.

```python
# For dataset1
spectral_features_dataset1 = []
for i in range(dataset1.shape[0]):
    spectral_features_dataset1.append(get_spectral_features(dataset1.iloc[i, :], d1samprate))
dataset1['spectral_features'] = spectral_features_dataset1

# For dataset2
spectral_features_dataset2 = []
for i in range(dataset2.shape[0]):
    spectral_features_dataset2.append(get_spectral_features(dataset2.iloc[i, :], d2samprate))
dataset2['spectral_features'] = spectral_features_dataset2
```

### Descriptive Statistics
Here we will seperate response frequencies into delta, theta, alpha, beta, gamma waves for trials in both the seizure and non-seizure groups. We first do this for the last five data samples for both sets to see how or if they differ from eachother.

```python
# Define EEG bands
eeg_bands = {'Delta': (0.5, 4),
             'Theta': (4, 8),
             'Alpha': (8, 12),
             'Beta': (12, 30),
             'Gamma': (30, 100)}
def get_band_powers(spectral_data, freqs):
    # Initialize the band power dictionary
    band_powers = {}

    # Calculate the power of each frequency band
    for band in eeg_bands: 
        low, high = eeg_bands[band]
        # Use boolean mask to get frequencies in the right range
        mask = (freqs > low) & (freqs <= high)
        band_power = spectral_data[mask].mean(0)
        band_powers[band] = band_power
    return band_powers
    return band_powers

# For dataset1
band_powers_dataset1 = []
for i in range(-5, 0):  # Last five samples
    freqs, psd = welch(dataset1.iloc[i], fs=d1samprate, nperseg=d1samprate)
    band_powers_dataset1.append(get_band_powers(psd, freqs))

# For dataset2
band_powers_dataset2 = []
for i in range(-5, 0):  # Last five samples
    freqs, psd = welch(dataset2.iloc[i], fs=d2samprate, nperseg=d2samprate)
    band_powers_dataset2.append(get_band_powers(psd, freqs))
    
band_powers_dataset1_df = pd.DataFrame(band_powers_dataset1)
band_powers_dataset2_df = pd.DataFrame(band_powers_dataset2)

print("Dataset 1")
print(band_powers_dataset1_df.mean())
print(band_powers_dataset1_df.std())
print("\nDataset 2")
print(band_powers_dataset2_df.mean())
print(band_powers_dataset2_df.std())
```

# Feature Extraction

Here we are using the control trials only to create a ' Normal Operating Range' of brain function, this will be used to detect seizures in our trials.

```python
# first we need to plot out average amplitude at every sample
# then we need to create bounding range to cover ~95% of the data
controltrials = clean1[clean1['y'] == 0]
seizuretrials = clean1[clean1['y'] == 1]

controltrials = controltrials.drop(columns = 'y')
control1d = np.array(controltrials).flatten()
control1d.shape
```

```python
avgsamp = control1d.mean()
stdsamp = control1d.std()
print(avgsamp, stdsamp)
```

```python
normalrange = []
normalrange.append(avgsamp - 1.96* stdsamp)
normalrange.append(avgsamp + 1.96* stdsamp)
print(normalrange)
```

The red lines indicate our normal range outlined in the cell above.

```python
plt.plot(data1.loc[0], color = 'orange')
plt.plot(data1.loc[1], color = 'blue')
plt.axhline(y = normalrange[0], color = 'red')
plt.axhline(y = normalrange[1], color = 'red')
plt.xticks(visible = False)
plt.legend(['Seizure','No Seizure'])
plt.ylabel('Amplitude')
plt.xlabel('Samples')
plt.title('Normal Operating EEG Amplitude');
```

We will now use this to calculate the percent of each trial spent outside of the normal operating range, giving us our "NOR" extracted feature.

```python
def featextract(trial):
    if len(trial.shape) > 1:
        sr = trial.shape[1]
#         print(sr) 
    
    elif len(trial.shape) == 1:
        sr = len(trial)
        count = 0
        for x in range(sr):
            if  trial[x] < normalrange[0] or trial[x] > normalrange[1]:
                count += 1
        return count/sr # returns percentage of samples outside of normal operating range as a value between 0 and 1
#         print(sr)
        
featextract(data1.loc[0])
```
